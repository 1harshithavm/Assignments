```python
#Assignment(28/02/2026)
#Assignment Name : Storytelling with graphs

import matplotlib.pyplot as plt

# Sample Data
students = ["Asha", "Rahul", "Kiran", "Meena", "Arjun"]
marks = [90, 78, 85, 67, 54]

# -----------------------------
# 1️⃣ Bar Chart
# -----------------------------
plt.figure()
plt.bar(students, marks)
plt.title("Student Marks - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# -----------------------------
# 2️⃣ Pie Chart
# -----------------------------
plt.figure()
plt.pie(marks, labels=students, autopct="%1.1f%%")
plt.title("Marks Distribution - Pie Chart")
plt.show()

# -----------------------------
# 3️⃣ Histogram
# -----------------------------
plt.figure()
plt.hist(marks, bins=5)
plt.title("Marks Distribution - Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#assignment (17/02/2026)
#assignment Name:Logic Builder


def fizz_buzz_builder(limit):
    counts = {"Fizz": 0, "Buzz": 0, "FizzBuzz": 0, "Numbers": 0}

    for i in range(1, limit + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
            counts["FizzBuzz"] += 1
        elif i % 3 == 0:
            print("Fizz")
            counts["Fizz"] += 1
        elif i % 5 == 0:
            print("Buzz")
            counts["Buzz"] += 1
        else:
            print(i)
            counts["Numbers"] += 1

    return counts

# Execution
results = fizz_buzz_builder(50)
print(f"\nSummary: {results}")
```

```python
#Assignment(19/02/2026)
#Assignment Name:Student data manager

students = {
    "Asha": 92,
    "Rahul": 85,
    "Kiran": 74,
    "Meena": 66,
    "Arjun": 58
}
total_marks = 0
highest_marks = 0
topper = ""

for name, marks in students.items():
    total_marks += marks

    if marks > highest_marks:
        highest_marks = marks
        topper = name

print("Topper of the class:", topper, "with marks", highest_marks)

average = total_marks / len(students)
print("Class Average:", average)

print("\nStudent Grades:")

for name, marks in students.items():

    if marks > 90:
        grade = "A+"
    elif marks > 80:
        grade = "A"
    elif marks > 70:
        grade = "B+"
    elif marks > 60:
        grade = "B"
    else:
        grade = "C"

    print(name, "-", marks, ":", grade)
```

```python
import bcrypt

def hash_password(password):
    # Generate a salt and hash the password
    # gensalt() generates a new salt for each password
    hashed_bytes = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())
    return hashed_bytes.decode('utf-8') # Store the hash as a string

def check_password(password, hashed_password):
    # Check if the provided password matches the stored hash
    return bcrypt.checkpw(password.encode('utf-8'), hashed_password.encode('utf-8'))


# --- Example Usage ---

# 1. When a user signs up or changes their password:
user_password = input("Enter a password to hash: ")
stored_hash = hash_password(user_password)
print(f"Password entered: {user_password}")
print(f"Stored hash (this is what you'd save in your database): {stored_hash}")

# 2. When a user tries to log in:
login_attempt_password = input("Enter password for login attempt: ")

if check_password(login_attempt_password, stored_hash):
    print("Login successful!")
else:
    print("Invalid password.")

# Try with a wrong password
print("\n--- Testing with a wrong password ---")
wrong_password = "wrongpass"
print(f"Attempting login with: {wrong_password}")
if check_password(wrong_password, stored_hash):
    print("Login successful!")
else:
    print("Invalid password.")
```
